```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/Lemonade.csv')
df.tail(10)
```

```python
# Here temp, rainfall and flyers are input features
# sales is out output feature
# split your data into training(67%) and testing part(33%)
# Apply multi linear regression algorithm on training data
# Predict for testing data
# evaluate the test data with predicted data using MSE(meam squared error)
```

```python
#input and output variables
x = df[['Temperature', 'Rainfall', 'Flyers']].values
y = df['Sales'].values.reshape(-1,1)
x,y
```

```python
#Equation for multi linear regression  equation
# y = m1x1+m2x2+m3x3+..........+mnxn + C
plt.scatter(df['Temperature'],df['Sales'], label = 'Temp')
plt.scatter(df['Rainfall'], df['Sales'], label = 'Rainfall')
plt.scatter(df['Flyers'], df['Sales'], label = 'Flyers')
plt.legend()

plt.show()
```

```python
#Now we will train data by using sklearn library
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


# Assume X is your features and y is your target variable
# Split the data: 67% for training and 33% for testing
X1_train, X1_test, y1_train, y1_test = train_test_split(x, y, test_size=0.33, random_state=42)

# Print the shapes of the resulting datasets
print("Training features shape:", X1_train.shape)
print("Testing features shape:", X1_test.shape)
print("Training target shape:", y1_train.shape)
print("Testing target shape:", y1_test.shape)

reg = LinearRegression().fit(X1_train,y1_train)

```

```python
#Predicted Data
y_pred = reg.predict(X1_test)
y_pred

```

```python
#Mean Squared Error
mse = mean_squared_error(y1_test, y_pred)
mse
```

```python
# Calculate accuracy
accuracy = reg.score(X1_test, y1_test)

# Print accuracy as percentage
print(f"Accuracy: {accuracy*100:.2f}%")
```

```python
#intercept
c = reg.intercept_
c
```
